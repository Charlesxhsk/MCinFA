## Brownian path simulation

```python
# a simple way to generate the path
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(200)
T = 1
N = 500
dt = T/N
dw = np.zeros(N)
w = np.zeros(N)

dw[1] = np.sqrt(dt) * np.random.normal(0,1)
for i in range(2,N):
    dw[i] = np.sqrt(dt) * np.random.normal(0,1)
    w[i] = w[i-1] + dw[i]

plt.plot(np.arange(0,T,dt),w)
plt.xlabel('t')
plt.ylabel('w(t)')
plt.title('Brownian path simulation 1')
```

```python
# use vectorized method to generate path
np.random.seed(200)
T = 1
N = 500
dt = T/N

dw = np.sqrt(dt) * np.random.normal(0,1,N-1)
w = np.cumsum(dw)
w = np.insert(w,0,0)

plt.plot(np.arange(0,T,dt),w)
plt.xlabel('t')
plt.ylabel('w(t)')
plt.title('Brownian path simulation 2')
```
